# Phase 1: Market Data Ingestion & Exploration

```python
import yfinance as yf
import pandas as pd

tickers = [
    "RELIANCE.NS", "TCS.NS", "INFY.NS", "HDFCBANK.NS", "ICICIBANK.NS",
    "ITC.NS", "LT.NS", "KOTAKBANK.NS", "SBIN.NS", "AXISBANK.NS",
    "HINDUNILVR.NS", "BAJFINANCE.NS", "BHARTIARTL.NS", "MARUTI.NS", "HCLTECH.NS",
    "SUNPHARMA.NS", "TITAN.NS", "ULTRACEMCO.NS", "NESTLEIND.NS", "WIPRO.NS"
]  # your list of tickers
start = "2018-01-01"
end = "2023-12-31"

# ✅ Initialize list to store data
all_data = []

# ✅ Download for each stock
for ticker in tickers:
    print(f"Fetching {ticker}...")
    df = yf.download(ticker, start=start, end=end)
    df["Ticker"] = ticker
    df.reset_index(inplace=True)
    all_data.append(df)  # ✅ appending DataFrame to list

# ✅ Combine and save
full_df = pd.concat(all_data, ignore_index=True)
full_df.to_csv("/content/drive/MyDrive/algo_trading_project/data/nifty50_data.csv", index=False)
print("✅ Done! Saved to /content/drive/MyDrive/algo_trading_project/data/nifty50_data.csv")

# ✅ Preview
full_df["Ticker"].value_counts().head()
```
